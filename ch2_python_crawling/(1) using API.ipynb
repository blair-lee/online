## 1. requests 모듈 사용하기(HTTP 통신)

**requests 모듈**
1. http request/response를 위한 모듈
2. HTTP method를 메소드 명으로 사용하여 request 요청 ... ex)get, post

```python
import requests
```

**get 요청하기**
1. http get 요청하기
2. query parameter 이용하여 데이터 전달하기

```python
url = 'https://news.v.daum.net/v/20210204030153200'
resp = requests.get(url)

resp.text
```

**post 요청하기**
1. http post 요청하기
2. post data 이용하여 데이터 전달하기

```python
url = 'https://www.kangcom.com/member/member_check.asp'
data = {
    'id': 'macmath22',
    'pwd': 'T'
}

resp = requests.post(url, data=data)
resp.text
```
