```python
# ì†Œìˆ˜ íŒë³„ (1ê³¼ ìê¸° ìì‹ ìœ¼ë¡œë§Œ ë‚˜ëˆ ì§€ëŠ” ìˆ˜)
# ì…ë ¥: ì–‘ì˜ ì •ìˆ˜ 1ê°œ
# ì¶œë ¥: boolean (ì†Œìˆ˜: True, í•©ì„±ìˆ˜: False)

# 16
# 2, 3, 4, 5, 6, 7, 8, .... 15
def is_prime(num):
    for i in range(2, num):
        if num % i == 0: # ë‚˜ëˆ  ë–¨ì–´ì§€ë©´ 
            return False
    return True  
    
print(is_prime(100))
print(is_prime(89))
print(is_prime(17))
print(is_prime(2))
```

```python
# 2, 3, 4, 5, 6, 7 -> 4
# 2, 3, 4, 5 -> 3
# ì…ë ¥: ì–‘ì˜ ì •ìˆ˜ 1ê°œ
# ì¶œë ¥: 2-í•´ë‹¹ ìˆ«ì ì‚¬ì´ì˜ ì†Œìˆ˜ì˜ ê°œìˆ˜

def num_prime(num):
    count = 0
    for i in range(2, num+1):
        if is_prime(i):
            count += 1
    return count

print(num_prime(7))
print(num_prime(5))
print(num_prime(100))
```
